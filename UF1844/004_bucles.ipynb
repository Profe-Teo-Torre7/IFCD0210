```python
lista = [1,2,3]
otra = [4,5,6]
total = lista + otra

print(total)
```

```python
for n in total:
    if n % 2 == 0:
        print(n)
    else:
        print('-')
```

# Bucles while

```python
contador = 0
while contador < 10:
    print(contador)
    contador += 1

print("Fuera del bucle")
```

```python
final = 10
control = 0

while True:
    print(f"Hola {control}")
    if control == final:
        break
    control += 1
```
