```python
dato = input("Dame un numerito")

numero = int(dato)
if numero > 0:
    print(f'El número {numero} es mayor que cero')
elif numero == 0:
    print(f'El número {numero} es cero')
else:
    print(f'El número {numero} es menor que cero')

```
